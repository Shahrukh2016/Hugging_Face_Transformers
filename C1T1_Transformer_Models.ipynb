### Model1: sentiment-analysis



```python
## Importing pipeline library from transformers module and creating instance for model sentiment-analysis
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
### Testing classifier (Model sentiment-analysis, Test 1)
classifier("I don't want to go there!")
```

```python
### Testing classifier (Model sentiment-analysis, Test 2)
classifier(["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!", "Perfect"])
```

### Model2: zero-shot-classification

```python
## Importing pipeline library from transformers module and creating instance for model zero-shot-classification
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
```

```python
### Testing classifier (Model zero-shot-classification, Test 1)
classifier("This is a course about the Transformers library", candidate_labels=["education", "politics", "business"])
```

```python
### Testing classifier (Model zero-shot-classification, Test 2)
classifier("Stocks are rising", candidate_labels=["education", "weather", "business"])
```
